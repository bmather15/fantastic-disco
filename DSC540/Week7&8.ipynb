```python
#Filling in missing data with 0's
data.fillna(0)
```

```python
#Filling in missing data for Q3: AGE column so I can bin them
data['Q3: AGE'] = (pd.to_numeric(data['Q3: AGE'], errors='coerce').fillna(0))
```

```python
#Binning into age groups
bins = [1, 18, 25, 30, 40, 50, 60, 70, 100]
```

```python
age_bins = pd.cut(data['Q3: AGE'], bins)
```

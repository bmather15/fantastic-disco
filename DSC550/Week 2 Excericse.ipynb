```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Downloading CSV
data = pd.read_html('https://www.pro-football-reference.com/years/2022/fantasy-points-against-QB.htm#fantasy_def')[0]
```

```python
data.columns = data.columns.droplevel()
```

```python
data = data.drop(['FantPt', 'DKPt', 'FDPt'], axis=1)
```

```python
data.columns = ['Team','Games','Passing Completions','Passing Attempts','Passing Yards Allowed','Passing Touchdowns',
                   'Interceptions','Two Point Conversions','Sacks', 'Rushing Attempts', 'Rushing Yards Allowed', 'Rushing Touchdowns']
```

```python
data.fillna(0)
```

The question I would like to explore is what affect Interceptions and Sacks have on the QB, weather it will make them pass/run better or worse. 

3) Create at least three graphs that help answer these questions. Make sure your graphs are clearly readable and are labeled appropriately and professionally.

```python
print("Interceptions/Sacks vs Rushing Attempts: ")
x=data["Rushing Attempts"]
y=data["Sacks"]
plt.scatter(x, y, c='green')

x=data["Rushing Attempts"]
y=data["Interceptions"]
plt.scatter(x, y, c='red')
plt.xlabel("Rushing Attempts")
plt.ylabel("Interceptions/Sacks")
plt.legend(["Sacks" , "Interceptions"])
plt.show()
```

```python
print("Interceptions/Sacks vs Passing Attempts: ")
x=data["Passing Attempts"]
y=data["Sacks"]
plt.scatter(x, y, c='green')

x=data["Passing Attempts"]
y=data["Interceptions"]
plt.scatter(x, y, c='red')
plt.xlabel("Passing Attempts")
plt.ylabel("Interceptions/Sacks")
plt.legend(["Sacks" , "Interceptions"])
plt.show()
```

```python
x=data["Passing Yards Allowed"]
z=data["Rushing Yards Allowed"]
y=data["Interceptions"]
w=data["Sacks"]
```

```python
print("Interceptions vs Passing Yards Allowed: ")
plt.plot(y, x)
plt.xlabel("Interceptions")
plt.ylabel("Passing Yards Allowed")
plt.show()
```

```python
print("Sacks vs Passing Yards Allowed: ")
plt.plot(w, x)
plt.xlabel("Sacks")
plt.ylabel("Passing Yards Allowed")
plt.show()
```

```python
print("Interceptions vs Rushing Yards Allowed: ")
plt.plot(y, z)
plt.xlabel("Interceptions")
plt.ylabel("Rushing Yards Allowed")
plt.show()
```

```python
print("Sacks vs Rushing Yards Allowed: ")
plt.plot(w, z)
plt.xlabel("Sacks")
plt.ylabel("Rushing Yards Allowed")
plt.show()
```

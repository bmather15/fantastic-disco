2) Create a histogram or bar graph from your data.


```python
histogram = sns.histplot(data = stats, x="PF", bins=32)
histogram.set(xlabel='Points Scored',
       ylabel='Amount of Teams',
       title='Points Scored in 2022 Season')
```

3) Create a boxplot from your data. 

```python
plot = sns.boxplot(x='W', y='SoS', data=stats)
plot.set(xlabel='Wins',
       ylabel='Strength of Schedule',
       title='Strength of Schedule vs Wins in 2022 Season')
```

4) Create a bivariate plot from your data.

```python
scatter = sns.scatterplot(data=stats, x="W", y="PF")
scatter.set(xlabel='Wins',
       ylabel='Points Scored',
       title='Points Scored vs Wins in 2022 Season')
```

5) Create any additional visualizations that will help to answer the question(s) you want to answer.

```python
scatter2 = sns.scatterplot(data=stats, x="W", y="PA")
scatter2.set(xlabel='Wins',
       ylabel='Points Allowed',
       title='Points Allowed vs Wins in 2022 Season')
```

```python
scatter3 = sns.scatterplot(data=stats, x="SoS", y="PF")
scatter3.set(xlabel='Strength of Schedule',
       ylabel='Points Scored',
       title='Points Scored vs Strength of Schedule in 2022 Season')
```

```python
scatter4 = sns.scatterplot(data=stats, x="SoS", y="PA")
scatter4.set(xlabel='Strength of Schedule',
       ylabel='Points Allowed',
       title='Points Allowed vs Strength of Schedule in 2022 Season')
```
